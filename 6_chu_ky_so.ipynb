# 1. Khái niệm chữ ký số
- Chữ ký số là một dạng chữ ký điện tử được tạo ra bằng sự biến đổi một thông điệp dữ liệu sử dụng hệ thống mật mã không đối xứng, theo đó, người có được thông điệp dữ liệu ban đầu và khóa công khai của người ký có thể xác định được chính xác: a) Việc biến đổi nêu trên được tạo ra bằng đúng khóa bí mật tương ứng với khóa công khai trong cùng một cặp khóa; b) Sự toàn vẹn nội dung của thông điệp dữ liệu kể từ khi thực hiện việc biến đổi nêu trên.
# 2. Ưu điểm của chữ ký số
- Tiết kiệm thời gian, nguồn lực trong quá trình vận hành, làm việc nội bộ và với khách hàng.
- Khẳng định trách nhiệm pháp lý, tính bằng chứng chứng cứ của tài liệu sau ký
- Đảm bảo tính toàn vẹn dữ liệu
# 3. Chữ ký số RSA
- Quá trình tạo chữ ký số bao gồm hai phần chính:
  + Khóa công khai (Public Key): Được sử dụng để xác minh chữ ký số. Khóa này có thể được phân phối rộng rãi.
  + Khóa riêng (Private Key): Được sử dụng để tạo chữ ký số. Khóa này được giữ bí mật bởi người ký.
- Quá trình ký số diễn ra như sau:
  + Người ký sử dụng khóa riêng của mình để tạo ra chữ ký số từ dữ liệu gốc.
  + Chữ ký số này cùng với dữ liệu gốc được gửi đến người nhận.
  + Người nhận sử dụng khóa công khai của người ký để xác minh chữ ký số, đảm bảo rằng dữ liệu không bị thay đổi và thực sự đến từ người ký.

## 3.1. Tạo chữ ký số
- Băm dữ liệu: Dữ liệu gốc được chuyển đổi thành một chuỗi băm cố định (hash) sử dụng một hàm băm như SHA-256.
- Mã hóa chuỗi băm: Chuỗi băm được mã hóa bằng khóa riêng của người ký, tạo thành chữ ký số.
- Gửi dữ liệu và chữ ký: Dữ liệu gốc và chữ ký số được gửi đến người nhận.

## 3.2. Xác minh chữ ký số
- Giải mã chữ ký số: Người nhận sử dụng khóa công khai của người ký để giải mã chữ ký số, thu được chuỗi băm gốc.
- Băm lại dữ liệu: Dữ liệu gốc nhận được được băm lại sử dụng cùng hàm băm.
- So sánh hai chuỗi băm: So sánh chuỗi băm nhận được từ chữ ký số và chuỗi băm tính lại từ dữ liệu gốc. Nếu chúng khớp nhau, chữ ký số là hợp lệ.

## 3.3. Ví dụ
Thông điệp (Message): M = "Hello"
- Khóa RSA:

  Private key (d, n): (29, 35)

  Public key (e, n): (5, 77)
  
- Hàm băm SHA-256: H(M) = 185f8db32271fe25f561a6fc938b2e264306ec304eda518007d1764826381969

  Do H(M_ là số nguyên lớn. Giả sử giá trị băm của thông điệp "Hello" là H(M) = 32.
- Mã hóa hàm băm bằng khóa private (Sign the Hash):

  Dùng khóa private (d, n) để mã hóa giá trị băm:

  Signature = (H(M)^d) mod n = (32^29) mod 35 = 2

- Xác thực chữ ký (Verify the Signature):

  Dùng khóa public (e, n) để xác thực chữ ký:
  
  Recovered Hash = (Signature^e) mod n = (2^5) mod 35 = 32

- So sánh giá trị băm gốc với giá trị băm thu được (Compare the Hashes):

  Giá trị băm gốc H(M) = 32

  Giá trị băm thu được Recovered Hash = 32

  => Chữ ký hợp lệ.

```python
# Function to find gcd 
# of two numbers
def euclid(m, n):
	
	if n == 0:
		return m
	else:
		r = m % n
		return euclid(n, r)
	
	
# Program to find 
# Multiplicative inverse
def exteuclid(a, b):
	
	r1 = a
	r2 = b
	s1 = int(1)
	s2 = int(0)
	t1 = int(0)
	t2 = int(1)
	
	while r2 > 0:
		
		q = r1//r2
		r = r1-q * r2
		r1 = r2
		r2 = r
		s = s1-q * s2
		s1 = s2
		s2 = s
		t = t1-q * t2
		t1 = t2
		t2 = t
		
	if t1 < 0:
		t1 = t1 % a
		
	return (r1, t1)

# Enter two large prime
# numbers p and q
p = 823
q = 953
n = p * q
Pn = (p-1)*(q-1)

# Generate encryption key 
# in range 1<e<Pn
key = []

for i in range(2, Pn):
	
	gcd = euclid(Pn, i)
	
	if gcd == 1:
		key.append(i)


# Select an encryption key 
# from the above list
e = int(313)

# Obtain inverse of 
# encryption key in Z_Pn
r, d = exteuclid(Pn, e)
if r == 1:
	d = int(d)
	print("decryption key is: ", d)
	
else:
	print("Multiplicative inverse for\
	the given encryption key does not \
	exist. Choose a different encryption key ")


# Enter the message to be sent
M = 19070

# Signature is created by Alice
S = (M**d) % n

# Alice sends M and S both to Bob
# Bob generates message M1 using the
# signature S, Alice's public key e 
# and product n.
M1 = (S**e) % n

# If M = M1 only then Bob accepts
# the message sent by Alice.

if M == M1:
	print("As M = M1, Accept the\
	message sent by Alice")
else:
	print("As M not equal to M1,\
	Do not accept the message\
	sent by Alice ")
```
